# Importing and Reading Datasets

The basic importing and reading the dataset includes:

1. Importing the library and the dataset
2. Viewing dimension and shape
2. Viewing the top/bottom-n results with `head(n)` and `tail(n)` functions
3. Viewing data with all the columns displayed

## Importing the library and the dataset

```python
import pandas as pd
```

```python
data = pd.read_csv('../00_Datasets/Toyota.csv')
data
```

## Dimensionality of the Dataset

```python
# Viewing the dimension of the data :

data.ndim
```

```python
# Viewing the rows and columns of the data:

data.shape
```

```python
# Viewing the total datapoints (rows x columns):

data.size
```

## Viewing Portion of Dataset

```python
# Viewing first 10 rows of data:

data.head(10)
```

```python
# Viewing last 10 rows of data:

data.tail(10)
```

## Viewing all the columns and rows

```python
# settings to view all the columns:

pd.set_option('display.max_columns', 11)
data
```

```python
# settings to view all n-rows:

pd.set_option('display.max_rows', 1436)
data
```
